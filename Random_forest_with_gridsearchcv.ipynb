Problem Statement 3: Load the ‘voice.csv’ dataset into a DataFrame and perform the following
tasks:
1. Considering the ‘label’ column as the target variable, rename the column as
‘Gender_Identified’
2. Using the preprocessing() function, label the target column
3. Separate the target variable and the feature vectors
4. Build a RandomForestClassifier model and find the best parameters using a Grid search
5. Print the best parameters and the best estimator

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv("voice.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.rename(columns = {'label':'Gender_Identified'}, inplace = True)
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df.iloc[:,-1]= le.fit_transform(df.iloc[:,-1])
df.head()
```

```python
X = df.drop(columns = ['Gender_Identified'])
X.head()
```

```python
y = df['Gender_Identified']
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,y_test = train_test_split (X, y, test_size=.2, random_state = 100)
```

```python
rf = RandomForestClassifier()
rf.fit(X_train,y_train)
y_pred = rf.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
params= {'criterion':["gini", "entropy", "log_loss"],
          'max_depth': range(2,10),
          'max_features': ['auto', 'sqrt', 'log2'],  
          'n_estimators' : [50,100,150,200]
          }
```

```python
g = GridSearchCV( estimator = rf, param_grid= params,scoring='accuracy', cv=4)
g.fit(X_train, y_train)
pred = g.predict(X_test)
accuracy_score(y_test, pred)
```

```python
print("\n The best estimator across ALL searched params:\n",g.best_estimator_)
print("\n The best score across ALL searched params:\n",g.best_score_)
print("\n The best parameters across ALL searched params:\n",g.best_params_)
```
